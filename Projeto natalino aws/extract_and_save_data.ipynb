Extraindo dados da API do Google Books

```python
import requests


url = "https://www.googleapis.com/books/v1/volumes"

parametros = {"q": "data engineering", "maxResults": 10}

response = requests.get(url, parametros)

data = response.json()["items"]

data
```

Processando os dados

```python
import pandas as pd

books = []

for item in data:
    books.append({
        "title": item["volumeInfo"].get("title", "Sem título"),
        "authors": ", ".join(item["volumeInfo"].get("authors", ["Desconhecido"])),
        "published_date": item["volumeInfo"].get("publishedDate", None),
        "description": item["volumeInfo"].get("description", "Sem descrição")
    })

books
```

```python
df = pd.DataFrame(books)
df
```

Salvando em arquivo CSV

```python
df.to_csv("tabela_livros.csv", index=False)
```
